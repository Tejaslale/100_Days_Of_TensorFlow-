```python
import tensorflow as tf
```

```python
categories = tf.constant([0, 1, 2, 0, 1, 2])
```

```python
num_categories = 3
encoding_layer = tf.keras.layers.CategoryEncoding(
  num_tokens=num_categories, output_mode="one_hot"
)
encoded_data = encoding_layer(categories)
```

```python
categories2 = tf.constant([1,2,4,1,4,2,1,3,4,3])
encoding_layer2 = tf.keras.layers.CategoryEncoding(num_tokens=5, output_mode="one_hot") # categories start from 1, so num_tokens should be 5
encoded_data2 = encoding_layer2(categories2)
encoded_data2.numpy()
```

```python
categories = tf.constant (["apple", "banana", "cherry", "apple", "banana", "cherry"])
hashing_layer = tf.keras.layers.Hashing(num_bins=3)
hashed_data = hashing_layer(categories)
hashed_data.numpy()
```

```python
# Sample categorical data
categories = ['red', 'blue', 'green', 'blue', 'red']

# Category Encoding
category_encoding_layer = tf.keras.layers.CategoryEncoding(num_tokens=4)
encoded_categories = category_encoding_layer(tf.constant([0, 1, 2, 1, 0]))
print('Category Encoding:\n', encoded_categories.numpy())
```

```python
# Hashing
hashing_layer = tf.keras.layers.Hashing(num_bins=4)
hashed_categories = hashing_layer(tf.constant(categories))
print('Hashing:\n', hashed_categories.numpy())
```

```python
# String Lookup
string_lookup_layer = tf.keras.layers.StringLookup()
string_lookup_layer.adapt(categories)
lookup_indices = string_lookup_layer(tf.constant(categories))
print('String Lookup:\n', lookup_indices.numpy())
```

```python
# Integer Lookup
integer_lookup_layer = tf.keras.layers.IntegerLookup()
integer_lookup_layer.adapt([0, 1, 2])
integer_indices = integer_lookup_layer(tf.constant([0, 1, 2]))
print('Integer Lookup:\n', integer_indices.numpy())
```
