```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
X = np.array([5,6,7,8,9,10])
y = np.array([7,8,9,10,11,12])
```

```python
X = tf.cast(tf.constant(X),dtype=tf.float32)
y = tf.cast(tf.constant(y),dtype=tf.float32)
```

```python
plt.scatter(X,y)
```

# model1

```python
tf.random.set_seed(42)

model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,)),
    tf.keras.layers.Dense(50, activation=None),
    tf.keras.layers.Dense(1, activation=None)
])

model1.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.01))
model1.fit(X, y, epochs=100, verbose=0)

prediction1 = model1.predict(np.array([11.0]))
print(f"Model 1 Prediction: {prediction1}")
```

# model2

```python
tf.random.set_seed(42)

model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,)),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(1, activation=None)
])

model2.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.01))
model2.fit(X, y, epochs=100, verbose=0)

prediction2 = model2.predict(np.array([11.0]))
print(f"Model 2 Prediction: {prediction2}")
```

# model3
