```python
# #Apply the @tf. function decorator to your Python function
import tensorflow as tf

@tf.function
def my_function(x, y):
  return tf.add(x, y)

result = my_function(tf.constant(2), tf.constant(3))
print(result)
```

```python
# prompt: def multiply(a, b):
@tf.function
def multiply(a, b):
  return a * b

multiply = tf.function(multiply)
```

```python
result = multiply(tf.constant(2), tf.constant (3))
print (result)
```

```python
@tf.function
def simple_function(x):
  return x * 2

result = simple_function(tf.constant(5))
print(result)
```

```python
#@tf. function
@tf.function
def tf_simple_function(x):
  return simple_function(x)
```

```python
# prompt: result = tf_simple_function(tf.constant(2))
result = tf_simple_function(tf.constant(2))
print (result)
```

```python
# Define the complex function
def complex_function(x, y):
  return tf.sin(x) + tf.cos(y)

# Use tf.GradientTape to calculate gradients
x = tf.Variable(2.0)
y = tf.Variable(3.0)

with tf.GradientTape() as tape:
  z = complex_function(x, y)

grad_x, grad_y = tape.gradient(z, [x, y])

print("Gradient with respect to x:", grad_x.numpy())
print("Gradient with respect to y:", grad_y.numpy())
```

```python
def complex_function(x, y):
  return x**2 + y**2
```

```python
x = tf.Variable(2.0)
y = tf.Variable(3.0)

with tf.GradientTape() as tape:
  z = complex_function(x, y)

dz_dx = tape.gradient(z, x)
print(dz_dx)

try:
  dz_dy = tape.gradient(z, y)
except RuntimeError as e:
  print(e)
```

```python
with tf.GradientTape(persistent=True) as tape:
  z = complex_function(x, y)

dz_dx = tape.gradient(z, x)
print(dz_dx)

dz_dy = tape.gradient(z, y)
print(dz_dy)
```
